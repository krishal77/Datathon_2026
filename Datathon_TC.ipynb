```python

def classify_income_by_gdp(gdp):
    if pd.isna(gdp):
        return None
    elif gdp < 1136:
        return 'Low'
    elif gdp < 4466:
        return 'Lower-Middle'
    elif gdp < 13846:
        return 'Upper-Middle'
    else:
        return 'High'

# Create new corrected column
df['income_level_corrected'] = df['gdp_per_capita_usd'].apply(classify_income_by_gdp)

# Compare old vs new
comparison = pd.crosstab(df['income_level'], df['income_level_corrected'], margins=True)


# Show countries that changed
changed = df[df['income_level'] != df['income_level_corrected']].groupby('country_code').size().sort_values(ascending=False)

# Replace old with new
df['income_level_old'] = df['income_level']  # backup
df['income_level'] = df['income_level_corrected']
df.drop('income_level_corrected', axis=1, inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle
from google.colab import files


uploaded = files.upload()
df = pd.read_csv("cleaned_dataset.csv")

X = df.drop(columns=['health_risk_class'])
y = df['health_risk_class']

categorical_cols = X.select_dtypes(include=['object', 'category']).columns
print("Categorical columns:", categorical_cols.tolist())

X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)
print("Shape after encoding:", X_encoded.shape)

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)


rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features='sqrt',
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)


print("Training Random Forest...")
rf_model.fit(X_train, y_train)


y_pred = rf_model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"\nRandom Forest Accuracy: {accuracy:.4f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))


with open('random_forest_model.pkl', 'wb') as f:
    pickle.dump(rf_model, f)

files.download('random_forest_model.pkl')
```

```python



import pandas as pd
import pickle
from google.colab import files  # only in Colab



with open('random_forest_model.pkl', 'rb') as f:
    rf_model = pickle.load(f)


uploaded = files.upload()
test_df = pd.read_csv("test.csv")


record_ids = test_df['record_id'] if 'record_id' in test_df.columns else pd.Series(range(1, len(test_df)+1))


columns_to_drop = ['record_id', 'country_name', 'date', 'year', 'week']
X_test = test_df.drop(columns=[col for col in columns_to_drop if col in test_df.columns])


categorical_cols = X_test.select_dtypes(include=['object', 'category']).columns
X_test_encoded = pd.get_dummies(X_test, columns=categorical_cols, drop_first=True)

train_features = rf_model.feature_names_in_
for col in train_features:
    if col not in X_test_encoded.columns:
        X_test_encoded[col] = 0
X_test_encoded = X_test_encoded[train_features]


probs = rf_model.predict_proba(X_test_encoded)


submission = pd.DataFrame(probs, columns=rf_model.classes_)
submission.insert(0, 'record_id', record_ids)


column_mapping = {
    rf_model.classes_[0]: 'Low_Risk',
    rf_model.classes_[1]: 'Moderate_Risk',
    rf_model.classes_[2]: 'High_Risk',
    rf_model.classes_[3]: 'Critical_Risk'
}
submission = submission.rename(columns=column_mapping)


submission = submission[['record_id', 'Low_Risk', 'Moderate_Risk', 'High_Risk', 'Critical_Risk']]


print("\nâœ… Submission preview:")
print(submission.head())


submission.to_csv("submission.csv", index=False)
files.download("submission.csv")
```
